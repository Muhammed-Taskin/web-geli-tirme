```python
#web sitesi içindeki tüm linkleri kontrol eden bir uygulama yapıcaz
#spider da denir
import requests
from bs4 import BeautifulSoup

target_url = "https://atilsamancioglu.com/"

found_links = set() # set ile burada aynı linklerin alınmasını engelledik.

def make_request(url):
    response = requests.get(url)
    print(response.text)
    #html parsing(bir sitenin içindeki linkleri ayıklayıp istek atmaya denir)
    #beautiful soup kullanımı
    soup = BeautifulSoup(response.text,"html.parser")
    return soup
def crawl(url):
    links = make_request(url) #make_request fonksiyonu ile url linklerini alıyoruz.
    for link in links.find_all("a"):
        found_links.add(link.get("href"))# href başlıklı linklerin alınmasını sağladık.
    print(found_links)
    #recursive(fonksiyon içinde fonksiyon çalıştırmaktır.devamlı çalışıcak)
    
    if "#" in found_links: #found_links içindeki linklerin içinde # varsa
        found_links.remove("#")# remove et sil demektir
        for link in found_links:crawl(found_links)

crawl(target_url)





```
