```python
#crypto crazy project
import requests

# get :sunucu ile iletişime geçme yöntemimiz(veriyi okumak istediğimizde kullanırız. )
site = requests.get("https://api.binance.com/api/v3/ticker/price")# binance dan veri çektik.
# post: backende bir veri gönderirken veya veri tabanında değişiklik yapılacaksa kullanılır.
# put or patch: veriyi güncellemek için kullanılır.
# delete: veriyi silmek için kullanılır.
print(site)
#response 200 isteğin başarılı olduğunu gösterir. 404 ise hatadır.
# 300 ise yönlendirme yapar. 500 ise sunucu hatasıdır.
print(site.status_code)# başarılı ise değer 200 olur
print(site.text)#veriyi okumak için kullanılır.
print(site.json())
if site.status_code == 200:
    for crypto in site.json():
        print(crypto["symbol"],crypto["price"])
```

```python
#concurrency: (birden fazla işi aynı anda çalıştırmak için kullanılır.)
#paralelism :(çoklu işi aynı anda çalıştırmak için kullanılır.)

import threading# bu modül birden fazla işi aynı anda çalışmasını sağlar.
import requests# istek atma kütüphanesi
import time# zaman bilgisi için

def get_data_sync(urls):
    st =time.time()# kullanılan sunucunun zamanını alıyoruz.
    json_array = []# json array adında bos bir liste oluşturuyoruz.
    for url in urls:
        json_array.append(requests.get(url).json())
        # döngü ile url içindeki bilgileri json formatında alıyoruz.
    et = time.time()
    elapsed_time = et -st# bize işlemin bitme zamanını veriyor.
    print(f"geçen zaman: {elapsed_time} saniye")
# buradan sonra threading işlemleri ile yaptığımız uygulamayı hızlandırıcaz.
class ThreadingDowloader(threading.Thread):
# şuan kalıtımı kullanıyoruz.
# basitçe kalıtım bir üst sınıfın methodlarını başka bir sınıf içerisinde kullanabilmemizi sağlar.
    json_array= []# json array adında boş bir liste oluşturuyoruz.
    
    def __init__(self,url):
        super().__init__()# kalıtımı kullanabilmek için super init kullanılmalı.
        self.url = url
    def run(self):
        response = requests.get(self.url)
        self.json_array.append(response.json())
        return self.json_array


def get_data_threading(urls):# işlemleri bölüp çalıştırdığımız fonksiyon.
    st = time.time()
    thread = []
    for url in urls:
        t = ThreadingDowloader(url)
        t.start()# şuan run fonksiyonu çalışıyor.
        thread.append(t)# append ile threadleri ekliyoruz.
    for t in thread:
        t.join()
        print(t)
    et =time.time()
    elapsed_time = et -st
    print(f"geçen zaman: {elapsed_time} saniye")
    
urls = ["http://echo.getpostman.com/delay/3"]*10
        
get_data_threading(urls)# işlemi 40 snden 2.5 saniyeye düşürdük.
#asyncio threading işleminde farklı bir yöntem
import asyncio # bu kütüphane internetten çalışırken threading kütüphanesine göre daha güvenlidir
#async (senkron olmayan şekilde bir fonksiyon oluşturacağımızı belirtir.)
#await (bunun sonucunu bekleyeceğimizi belirtir.)
#gather (birden fazla fonksiyonu içinde çalıştırabileceğimiz bir yapıdır.)
async def get_data(session,url,json_array):
    async with session.get(url) as resp:
        json_array.append(await resp.json())
        
    







```

```python
import requests

url = requests.get("https://api.binance.com/api/v3/ticker/price")#burada binance apiye istek attık
def crypto2():# burada artık bir fonksiyon oluşturarak verileri json formatında aldık
    if url.status_code == 200:# eğer bağlantı başarılı ise json formatında fonksiyona sunacak.
        return url.json()
    else:
        "Bağlantı Başarısız"
kripto_istek = input("Lütfen Tüm Harfler Büyük Şekilde İstediğiniz Kriptoyu Giriniz(Yanına USDT yazınız.)").strip().upper()
#Kullanıcıdan istenilen coini aldık.
veriler = crypto2()# fonksiyonda varolan json formatındaki verileri veriler değişkenine eştiledik.
for veriler in veriler:# döngüler ile istenilen veriyi tek tek ele aldık.
    if kripto_istek == veriler["symbol"]:# burada ise kullanıcıdan alınan kriptonun verilerdeki sembol ile eşit olursa alttaki printi çalıştır demektir.
        print(f"{kripto_istek} fiyatı:{veriler["price"]}")
```
